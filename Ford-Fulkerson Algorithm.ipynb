```python
def BuildResidual(graph, flow):
    residualGraph = DiGraph()
    residualGraph.add_vertices(graph.vertices())    

    for u, v, cap in graph.edges(labels=True):       
        forwardCapacity = cap - flow[u][v]          
        if forwardCapacity > 0:                      
            residualGraph.add_edge(u, v, forwardCapacity)    
            
        backwardCapacity = flow[u][v]               
        if backwardCapacity > 0:                     
            if residualGraph.has_edge(v, u):         
                residualGraph.set_edge_label(v, u, temp_G.edge_label(v, u) + backwardCapacity)
            else:
                residualGraph.add_edge(v, u, backwardCapacity)
    return residualGraph
```

```python
def reconstructPath(parent, source, sink):
    path = []      
    node = sink       

    while parent[node] is not None:   
        prev = parent[node]           
        path.append((prev, node))    
        node = prev
    path.reverse()     
    return path
```

```python
def BreadthFirstSearch(resGraph, source, sink):
    visited = {source}              
    queue = deque([source])         
    parent = {source: None}         
    
    while queue:
        u = queue.popleft()        
        
        for v in resGraph.neighbor_out_iterator(u):  
            resCap = resGraph.edge_label(u, v)         
            if v not in visited and resCap > 0:      
                visited.add(v)                        
                parent[v] = u                       
                if v == sink:                         
                    return reconstructPath(parent, source, sink)   
                queue.append(v)    
    return None
```

```python
def FordFulkerson(G, s, t):
    flow = {u: {v: 0 for v in G.vertices()} for u in G.vertices()}

    resGraph = BuildResidual(G, flow) 
    resPath = BreadthFirstSearch(resGraph, s, t)

    maxFlow = 0
    
    while resPath is not None:
        delta = min(resGraph.edge_label(u, v) for (u, v) in resPath)   # checker for minimum weight or the bottleneck in the path
        
        for (u, v) in resPath:
            flow[u][v] += delta         # dis is da forwardCapacity update, when we found the delta. 
            flow[v][u] -= delta         # but in reverse, we also pull -- as per the algorithm steps
            
        maxFlow += delta          # actually, the delta is the safe flow we can put on any pipes per se, so if we sum it, we can assure that no pipe will explode given this sum
        
        # Checker/Information
        print(f"Flow augmented by {delta}, total max flow so far: {maxFlow}")
        print(resPath)
        for (u, v) in resPath:
            cap = resGraph.edge_label(u, v)
        
            if cap == delta:
                print("Bottleneck edge:", (u, v))
        
        
        print()
        resGraph = BuildResidual(G, flow)       # as long as there is a Path from the source to sink, we never stop, and this is actually the idea of max-flow/min-cut...
        resPath = BreadthFirstSearch(resGraph, s, t)    # since in the paper process of the algorithm, we delete or "cut" edges, then the graph orientation changes, and so is the s-t path

    return maxFlow     # is a number
```
