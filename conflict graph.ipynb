```python
# Independent Set
independentSets = set()

for v in G.vertices():
    I = set()
    candidateVertices = set(G.vertices())

    I.add(v)
    candidateVertices -= set(G.neighbors(v))
    candidateVertices.discard(v)

    while candidateVertices:
        arbitraryVertex = next(iter(candidateVertices))    # dis thing: next(iter(set), just iterates inside the setand find the 1st iliment
        I.add(arbitraryVertex)
        candidateVertices -= set(G.neighbors(arbitraryVertex))
        candidateVertices.discard(arbitraryVertex)
        
    independentSets.add(frozenset(I))     # simply, frozen set prevents the alteration of the set when another set is added so it become set of sets not a union of sets
    
print(independentSets)
```
